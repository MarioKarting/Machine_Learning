```python
import matplotlib.pyplot as plt
import random

#收集数据
x=range(60)
y=range(40)
#随机生成楚雄的天气,每分钟的天气情况都不同
y_chuxiong = [random.uniform(18,20) for i in x]
#构建画布
plt.figure(figsize=(20,8),dpi=100)

#生成折线图
plt.plot(x,y_chuxiong)

#构建刻度尺
x_ticks_lable=["11点{}分".format(i) for i in x]
y_ticks_lable = ["{}℃".format(i) for i in y]

#折线图对象刻度的调用 
plt.xticks(x[::5],x_ticks_lable[::5])
plt.yticks(y[::5],y_ticks_lable[::5])

#网格
plt.grid(True,linestyle = "--",alpha=0.5)

#添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("楚雄11点天气变化情况",fontsize = 20)


#保存图片必须在plt.show 之前
# plt.savefig("./chuxiong.png")

#展示画图
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

#收集数据
x=range(60)
y=range(40)
#随机生成楚雄的天气,每分钟的天气情况都不同
y_chuxiong = [random.uniform(18,20) for i in x]
#随机生成昆明的天气状况，每小时的
y_kunming=[random.uniform(22,24) for i in x]
#构建画布
plt.figure(figsize=(20,8),dpi=100)

#生成折线图
plt.plot(x,y_chuxiong,color="y",label="楚雄")
#生成昆明的折线图  
plt.plot(x,y_kunming,color="g",linestyle="--",label="昆明")

#构建刻度尺
x_ticks_lable=["11点{}分".format(i) for i in x]
y_ticks_lable = ["{}℃".format(i) for i in y]

#折线图对象刻度的调用 
plt.xticks(x[::5],x_ticks_lable[::5])
plt.yticks(y[::5],y_ticks_lable[::5])

#网格
plt.grid(True,linestyle = "--",alpha=0.5)

#添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("楚雄11点天气变化情况",fontsize = 20)


#保存图片必须在plt.show 之前
# plt.savefig("./chuxiong.png")

#显示图例的位置
plt.legend(loc="best")

#展示画图
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

#收集数据
x=range(60)
y=range(40)
#随机生成楚雄的天气,每分钟的天气情况都不同
y_chuxiong = [random.uniform(18,20) for i in x]
#随机生成昆明的天气状况，每小时的
y_kunming=[random.uniform(23,26) for i in x]
#构建画布
# plt.figure(figsize=(20,8),dpi=100)
fig,axes = plt.subplots(ncols=2,nrows=1,figsize=(20,8),dpi=100)

#绘制图像 
axes[0].plot (x,y_chuxiong,color="m",label="楚雄")
axes[1].plot (x,y_kunming,color="r",linestyle = "--",label="昆明")

#构建刻度尺
x_ticks_lable=["11点{}分".format(i) for i in x]
y_ticks_lable=["{}℃".format(i) for i in y]


axes[0].set_xticks(x[::5])
axes[0].set_yticks(y[::5])

axes[1].set_xticks(x[::5])
axes[1].set_yticks(y[::5])
#折线图对象刻度的调用 
# plt.xticks(x[::5],x_ticks_lable[::5])
# plt.yticks(y[::5],y_ticks_lable[::5])

axes[0].set_xticklabels(x_ticks_lable[::5])
axes[0].set_yticklabels(y_ticks_lable[::5])  

axes[1].set_xticklabels(x_ticks_lable[::5])
axes[1].set_yticklabels(y_ticks_lable[::5])  

#网格
axes[0].grid(True,linestyle = "--",alpha=0.5)
axes[1].grid(True,alpha=0.6)

#添加描述信息
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[0].set_title("楚雄11点天气变化情况",fontsize = 20)

axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")
axes[1].set_title("昆明11点天气变化情况",fontsize = 20)


#保存图片必须在plt.show 之前
# plt.savefig("./chuxiong.png")

#显示图例的位置
axes[0].legend(loc="best")
axes[1].legend(loc=0)


#展示画图
plt.show()
```
