```python
import pandas as pd
import numpy as np

pd.DataFrame(np.random.randn(2,3))

```

```python
#采集学生科目数据
score = np.random.randint(40,100,(10,5))
score_detail = pd.DataFrame(score)
score_detail
```

```python
score_detail.shape
```

```python
score_detail.shape[0]
```

```python
#构造行索引，特征索引 

subjects = ["语文", "数学", "英语", "政治", "体育"]
#构造列索引，样本-学生
stu = ["同学" + str(i) for i in range(score_detail.shape[0])]

#添加行列索引 

data = pd.DataFrame(score, columns=subjects, index=stu)
data
```

```python
arrays = [[1, 1, 2, 2], ['red', 'blue', 'red', 'blue']]
pd.MultiIndex.from_arrays(arrays, names=('number', 'color'))

# 结果
oo = pd.MultiIndex(levels=[[1, 2], ['blue', 'red']],
           codes=[[0, 0, 1, 1], [1, 0, 1, 0]],
           names=['number', 'color'])
oo
```
